1. Problem               
                                     


```python

sentence = input("Enter sentence: ")
sentence = sentence.lower()
word_list = sentence.split()

unique_words = []

for word in word_list:
    if word not in unique_words:
        unique_words.append(word)

# Printing results
print("Unique words count:", len(unique_words))
print("Unique words:", set(unique_words))
```

2. Problem

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_name = ""
highest_salary = 0

for name in employees:
    if employees[name] > highest_salary:
        highest_salary = employees[name]
        highest_name = name

print("Highest Salary:", highest_name, "-", highest_salary)
```

3. Problem

```python
numbers = [45, 22, 89, 10, 66]

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

4. Problem

```python
prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count = count + 1

print("Products above 1000:", count)
```

5. Problem

```python
attendance = ["P", "P", "A", "P", "P"]

present = 0

for day in attendance:
    if day == "P":
        present = present + 1

total_days = len(attendance)
percentage = (present / total_days) * 100

print("Attendance Percentage:", percentage)
```

7. Problem

```python
text = "pythonp"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] = frequency[char] + 1
    else:
        frequency[char] = 1

print(frequency)
```
