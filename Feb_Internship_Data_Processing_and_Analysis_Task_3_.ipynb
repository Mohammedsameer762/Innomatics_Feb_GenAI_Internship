Problem Statement 1: Employee Performance Bonus Eligibility


```python
# Employee performance data
employees = {"Ravi": 92,"Anita": 88,"Kiran": 92,"Suresh": 85}
highest_score = max(employees.values())

# Step 1: Find all employees with highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Step 2: Display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), "(Score:", highest_score, ")")
```

Problem Statement 2: Search Query Keyword Analysis


```python
# Input query
query = "Buy mobile phone buy phone online"
query = query.lower()

# Remove punctuation (basic method)
query = query.replace(".", "").replace(",", "")

# Split into words
words = query.split()

# Count frequency
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1
        # Show only words searched more than once
result = {}

for word, count in frequency.items():
    if count > 1:
        result[word] = count

print(result)
```

Problem Statement 3: Sensor Data Validation


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index in range(len(sensor_readings)):
    value = sensor_readings[index]

    if value % 2 == 0:   # Check if even
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Problem Statement 4: Email Domain Usage Analysis


```python
emails = ["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"]

domain_count = {}

# Count domains
for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Calculate percentage
total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(domain + ":", str(int(percentage)) + "%")
```

Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average
total = 0
for s in sales:
    total += s

average = total / len(sales)

# Step 2: Detect spikes (30% above average)
threshold = average * 1.3

for i in range(len(sales)):
    if sales[i] > threshold:
        print("Day", i + 1, ":", sales[i])
```

Problem Statement 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count_ids = {}

# Count occurrences
for user in user_ids:
    if user in count_ids:
        count_ids[user] += 1
    else:
        count_ids[user] = 1

# Display duplicates only
for user, count in count_ids.items():
    if count > 1:
        print(user, "→", count, "times")
```
